```python
sns.jointplot(x=tips['total_bill'],y=tips['tip'],color = 'k',size = 8,stat_func=pearsonr,kind='hex',space = 0. #hex ：蜂巢图（六边形图）
              ,ratio = 5)
```

```python
g = sns.jointplot(x='total_bill',y='tip',data = tips,stat_func=pearsonr,kind='kde',height = 8,space = 0. #kde ：密度图 , 阴影：shade_lowest = False
              ,ratio = 5)
g.plot_joint(plt.scatter, c = 'w', s = 10, linewidth = 1, marker='+') #在密度图的基础上绘制散点图
```

```python
import seaborn as sns
sns.set_style("white") #设置背景风格
msg = sns.load_dataset('tips')

#创建一个绘图区域
g = sns.JointGrid(x = "total_bill",y = 'tip',data=msg) #配置数据 
g.plot_joint(plt.scatter,edgecolor='white',alpha=0.9,color='b') #设置框内散点图 #alpha 为颜色深度
plt.grid(linestyle = '--') #网格
g.ax_marg_x.hist(msg['total_bill'],color='b',alpha=.3) #X轴直方图
g.ax_marg_y.hist(msg['tip'],color='r',alpha=.3,orientation='horizontal') #Y轴直方图
```
