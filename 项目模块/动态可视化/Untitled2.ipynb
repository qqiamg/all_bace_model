```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import scipy.stats as sci
```

```python
tips = sns.load_dataset('tips')
print(tips.head(10))
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter",height=8) 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind="hex",height=8,stat_func=pearsonr)
```

```python
g = sns.JointGrid(x = 'total_bill', y = 'tip', data = tips)
g = g.plot_joint(plt.scatter, color = 'g', s = 40, edgecolor = 'white',alpha=.5) #绘制散点图
plt.grid(linestyle = '--')
g.plot_marginals(sns.distplot, kde = True, color = 'g')
```
